```python
df = pd.read_csv("csv_dir/body.csv")
df
```

```python
plt.figure(figsize = (15,4))

# Load Height Histogram
plt.subplot(1,3,1)
plt.hist(df["height"], bins = 20, label = "height", edgecolor = "black")
plt.title("Height")
plt.legend()

# Load Weight Histogram
plt.subplot(1,3,2)
plt.hist(df["weight"], color = "gray", bins = 20, label = "weight", edgecolor = "black")
plt.title("Weight")
plt.legend()

# Scatter
plt.subplot(1,3,3)
plt.scatter(df["height"], df["weight"], alpha = 0.6, color= "green", s = 45, edgecolor = "black")
plt.xlabel("height")
plt.ylabel("weight")
plt.title("Height & Weight Distribution")

# 공통 메소드 사용
plt.tight_layout()
plt.show()
```

```python
# 평균이 10, 표준편차가 1인 숫자를 100개 만들어라. 보통 플마 3정도나옴
data = np.random.normal(10,1,100)
plt.figure(figsize = (10,5))
# 20개의 구간으로 나눠서 density = 전체의 합이 1이 되도록 바꿈, density 안 쓰면 그 구간의 개수
plt.hist(data, bins = 20,density = True)
plt.title("histogram example")
plt.show()
```

```python
# 파이 차트
sizes = [25, 30, 20, 25] # 값
labels = ['A', 'B', 'C', 'D'] # 카테고리
colors = ['red', 'blue', 'green', 'yellow']
explodes = (0, 0.1, 0, 0) # 띄울 수 잇음
start_angle = 90 # 시작 각도조절
text_props = {'color':'white'} # 원 내부의 글자 색상
```

```python
plt.pie(sizes, labels=labels, colors = colors, explode = explodes, autopct='%1.1f%%', startangle=start_angle,counterclock=False, shadow=True, textprops=text_props)

plt.title("pie chart", fontsize = 20, fontweight = 'bold')
plt.xlabel("Label")
plt.ylabel("Alpha")
plt.tight_layout() # 안 잘리게 레이아웃 자동 최적화
plt.show()
```

```python
# 다중 레이아웃 subplot()
# figuresize, subplot(row,col,index), plot, subplot(row,col,index), tight_layout(), show() 순서대로 하기

x = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize = (10,5))


plt.subplot(1,2,1)
plt.plot(x,np.sin(x))
plt.title("sin")

plt.subplot(1,2,2)
plt.plot(x,np.cos(x))
plt.title("cos")



plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize = (5,5))
for i in range(0,9):
    plt.subplot(3,3,i+1)
    plt.plot(x, np.sin(x+i), label = f"{i}sin")
    plt.title(f"{i+1} sine")
    plt.yticks([])
    plt.legend()
plt.tight_layout()
plt.show()
```
