```python
import numpy as np
import pandas as pd

# 예시: df['Quantity']가 있다고 가정
# Q1, Q2, Q3 계산
q1 = np.percentile(df['Quantity'], 25)
q2 = np.percentile(df['Quantity'], 50)  # 중앙값
q3 = np.percentile(df['Quantity'], 75)

# IQR
iqr = q3 - q1

# 이상치 기준 (하한, 상한)
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print("하위 25% (Q1):", q1)
print("중앙값    (Q2):", q2)
print("상위 25% (Q3):", q3)
print("IQR:", iqr)
print("하한:", lower_bound)
print("상한:", upper_bound)

# 이상치 데이터
outliers = (df['Quantity'] < lower_bound) | (df['Quantity'] > upper_bound)

# 이상치 제거
# df_no_outliers = df[(df['Quantity'] >= lower_bound) & (df['Quantity'] <= upper_bound)]

# print("\n이상치 제거 후 데이터 개수:", len(df_no_outliers))
```
