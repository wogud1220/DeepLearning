```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
x = [5, 7, 8, 9, 10, 12, 13, 14, 21]
# IQR 구하기
q1 = np.percentile(x, 25)
q2 = np.percentile(x, 50)
q3 = np.percentile(x, 75)
iqr = q3 - q1
print("Q1", q1)
print("Q2 (중앙값)", q2)
print("Q3", q3)
print("IQR", iqr)
print(f"이상치: {q3 + iqr * 1.5} 이상")
print(f"이상치: {q1 + iqr * 1.5} 이하")
```

```python
#box plot



# plt.figure(figsize=(10, 20))
plt.boxplot(x,
            patch_artist=True, # 박스 색 채우기,
            # 박스 색 설정
            # boxprops = dict(facecolor='skyblue', color='lightblue'), #
            # 중앙선 설정
            medianprops = dict(color='red', linestyle='-', linewidth=4),
            # 가운데 선 설정
            whiskerprops = dict(color='black', linestyle='--'),
            # 이상치 모양 설정
            flierprops = dict(marker='*', markerfacecolor='red', markersize=5, color='red'),
            )
plt.title("사분위수 시각화", fontsize = 20, fontweight='bold')
plt.grid(linestyle='--')
# plt.yticks([1], ["A"])
plt.show()
```

```python
path = "./csv_dir/body.csv"
df = pd.read_csv(path)
df
```

```python
plt.figure(figsize = (12, 8))
sns.histplot(df,kde=True,bins=10,edgecolor='black')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = (10,5))

plt.subplot(1,3,1)
df['height'].plot(kind = 'kde')


plt.subplot(1,3,2)
sns.kdeplot(df['height'])

plt.subplot(1,3,3)


plt.show()
```

```python
plt.subplot(1,2,1)
plt.boxplot(df['height'], flierprops = dict(marker='^', markersize=4),
             medianprops = dict(color='red', linestyle='-', linewidth=4))
plt.grid(linestyle='--')
plt.title('height')
plt.ylabel('height')

plt.subplot(1,2,2)
plt.boxplot(df['weight'],
            flierprops = dict(marker='*', markersize=8),
             medianprops = dict(color='red', linestyle='-', linewidth=2),
            )
plt.grid(linestyle='--')
plt.title('weight')
plt.ylabel('weight')




plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = df['height'], y = df['weight'])
plt.title('height and weight')
plt.show()
```

```python
df['height_bin'] = pd.cut(df['height'], bins= range(150, 200, 5))
df['weight_bin'] = pd.cut(df['weight'], bins= range(40, 120, 10))
```

```python
plt.figure(figsize = (10,5))
sns.barplot(data = df, x= 'height_bin', y= 'weight')
plt.xticks(rotation=45)
plt.show()
# 150 초과 155개 이하
```

```python
plt.figure(figsize = (10,5))
sns.barplot(data = df, x= 'weight_bin', y= 'height')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.boxplot(data = df, x= 'weight_bin', y= 'height')
sns.boxplot(df['height'])
plt.xticks(rotation=20)
plt.show()
```

```python
# 바이올린 플ㄹ롯: boxplot + density

sns.violinplot(data = df, x= 'height_bin', y= 'weight')
plt.xticks(rotation=45)
plt.show()
```

```python
# 바이올린 플ㄹ롯: boxplot + density

sns.stripplot(data = df, x= 'height_bin', y= 'weight', jitter= 0.1)
plt.xticks(rotation=45)
plt.show()
```

```python
# joint plot


plt.figure(figsize = (10,5))
sns.jointplot(data = df, x= 'height', y= 'weight', kind="scatter")





plt.title("Joint Plot")
plt.show()
```
